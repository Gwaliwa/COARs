```python
# Loop through all TEXT files in a folder
import re
import os
all_files = os.listdir("C:/Users/ndehingia/Desktop/AI in eval/OOS Children Tami/COARs/TXTFiles/")
txt_files = filter(lambda x: x[-4:] == '.txt', all_files) 
```

```python
#create a dataframe where we will store the program priorities section
import pandas as pd
data0=pd.DataFrame(columns=['filename', 'context'])
```

```python
for i in range(n_file):
    txt_file=all_files[i]
    conclusions=" "
    with open('C:/Users/ndehingia/Desktop/AI in eval/OOS Children Tami/COARs/TXTFiles/'+ txt_file, encoding="utf8") as infile:
        copy = False
        for line in infile:
            #print(type(line))
           
           
            if "Update on the context and situation of children" in line.strip():
                copy = True
                continue
           
            if "Major contributions and drivers of results" in line.strip():
                copy = False
                continue
            
            elif copy:
                #outfile.write(line)
                conclusions=conclusions+line
    #data.iloc[i]=[txt_file,conclusions]
    data0 = data0.append({'filename' : txt_file, 'context' : conclusions},
        ignore_index = True)           
            
```

```python
#get the sentences before and after keywords appear:
#extract sentences before and after where keywords appear
list=['out-of-sch','out- of- sch','out - of - sch','out of school','school dropout','schooldrop','school retent','schoolretent', 'early warning systems','education management information system','earlywarningsystem','educationmanagementinformation','inclusive education','pro-poor equitable inclusive education','safe to learn','safetolearn','inclusiveeducat','safe-to-learn', 'digital learning', 'accessible digital textbook for all', 'learning passport', 'learningpassport', 'digitaltextbook','digitallearning', 'non-formal education','accelerated learning','short-term alternative education','catch-up program','bridging program','re-integration program','re-engagement program','alternative modalities of secondary education','technical and vocational education','skills development','technicaleducation','vocationaledu','skillsdevelop','non-formaledu','nonformaledu','acceleratedlearn','alternativeedu','bridgingprog']
import re
joined_word_list = '|'.join(map(re.escape, list))
```

```python
#number of sentences before and after the keyword that we want to extract
N=2
data['context_extracted_relevent_text'] = (data['context']
 .str.split('(?<=\.)\s*').explode()
 .loc[lambda d: 
      d.str.contains(joined_word_list)
       .groupby(level=0).rolling(2*N+1, center=True, min_periods=1).max()
       .droplevel(1).astype(bool)
     ]
 .groupby(level=0).agg(' '.join)
)
```
